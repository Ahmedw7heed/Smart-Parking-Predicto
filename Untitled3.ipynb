```python
"""
Project: Smart Parking & ANPR System
Institution: Al-Aqaba University for Technology (AUT)
Faculty: Artificial Intelligence and Data Engineering
Date: April 2026
"""

# تحميل النموذج الافتراضي
model = YOLO('yolov8n.pt')

def detect_vehicles_colab(img_path):
    # قراءة الصورة
    img = cv2.imread(img_path)

    if img is None:
        print("خطأ: تعذر العثور على الصورة.")
        return

    # تمرير الصورة لنموذج YOLO
    results = model(img)

    # رسم المربعات حول السيارات
    for result in results:
        boxes = result.boxes
        for box in boxes:
            x1, y1, x2, y2 = box.xyxy[0].cpu().numpy().astype(int)
            conf = float(box.conf[0])
            cls = int(box.cls[0])

            # فئة 2 تعني "سيارة"، ونشترط ثقة أعلى من 50%
            if cls == 2 and conf > 0.50:
                cv2.rectangle(img, (x1, y1), (x2, y2), (0, 255, 0), 2)
                cv2.putText(img, f'Car {conf:.2f}', (x1, y1 - 10),
                            cv2.FONT_HERSHEY_SIMPLEX, 0.9, (0, 255, 0), 2)

    # عرض الصورة باستخدام أداة Colab المخصصة
    cv2_imshow(img)

# استدعاء الدالة وتمرير الصورة التي رفعناها في الخلية الأولى
detect_vehicles_colab(image_path)
```

```python
"""
Project: Smart Parking & ANPR System
Institution: Al-Aqaba University for Technology (AUT)
Faculty: Artificial Intelligence and Data Engineering
Supervised By: Dr. Hassan Al-Rashaideh
Created By: Ahmed Wahed Abdel Aziz & Saber Gomaa Saber Oweis
Date: April 2026
"""

from google.colab import files
from ultralytics import YOLO
import easyocr
import cv2
import sqlite3
import numpy as np
from datetime import datetime
import pandas as pd
from google.colab.patches import cv2_imshow

print("⚙️ جاري تحميل النماذج وقاعدة البيانات...")

# 1. تهيئة قاعدة البيانات
conn = sqlite3.connect('smart_parking.db')
cursor = conn.cursor()
cursor.execute('''
CREATE TABLE IF NOT EXISTS entry_logs (
    id INTEGER PRIMARY KEY AUTOINCREMENT,
    plate_number TEXT NOT NULL,
    entry_time TIMESTAMP DEFAULT CURRENT_TIMESTAMP
)
''')
conn.commit()

# 2. تحميل نموذج YOLO وقارئ EasyOCR
model_path = '/content/runs/detect/aut_anpr_model-3/weights/best.pt'
custom_model = YOLO(model_path)
reader = easyocr.Reader(['en'], gpu=True) # تفعيل الإنجليزية للأرقام

print("\n📤 يرجى رفع صورة سيارة لاختبار النظام بالكامل...")
uploaded = files.upload()

if uploaded:
    test_image_path = list(uploaded.keys())[0]

    # 3. اكتشاف اللوحة
    results = custom_model(test_image_path)

    for result in results:
        # عرض الصورة الأصلية مع التحديد
        cv2_imshow(result.plot())

        boxes = result.boxes
        if len(boxes) == 0:
            print("⚠️ لم يتم اكتشاف لوحة.")

        for i, box in enumerate(boxes):
            # 4. قص اللوحة
            x1, y1, x2, y2 = box.xyxy[0].int().tolist()
            img = cv2.imread(test_image_path)
            cropped_plate = img[y1:y2, x1:x2]

            # 5. معالجة الصورة (تكبير + أبيض وأسود) لزيادة دقة القراءة
            enlarged = cv2.resize(cropped_plate, None, fx=2.5, fy=2.5, interpolation=cv2.INTER_CUBIC)
            gray = cv2.cvtColor(enlarged, cv2.COLOR_BGR2GRAY)
            _, thresh = cv2.threshold(gray, 100, 255, cv2.THRESH_BINARY | cv2.THRESH_OTSU)

            print("\n🔍 اللوحة بعد المعالجة لتوضيح الأرقام:")
            cv2_imshow(thresh)

            # 6. استخراج النص (أرقام فقط)
            ocr_results = reader.readtext(thresh, allowlist='0123456789', paragraph=False)
            plate_number = "".join([text for (bbox, text, prob) in ocr_results if prob > 0.4])

            # 7. الحفظ في قاعدة البيانات
            if plate_number:
                now = datetime.now().strftime("%Y-%m-%d %H:%M:%S")
                cursor.execute("INSERT INTO entry_logs (plate_number, entry_time) VALUES (?, ?)", (plate_number, now))
                conn.commit()
                print(f"✅ تمت العملية بنجاح! تم تسجيل السيارة: [{plate_number}]")
            else:
                print("⚠️ تم اكتشاف اللوحة لكن لم نتمكن من قراءة الأرقام بوضوح.")

    # 8. عرض التقرير النهائي
    print("\n📊 سجل النظام المحدث (Database Logs):")
    df = pd.read_sql_query("SELECT * FROM entry_logs ORDER BY id DESC LIMIT 5", conn)
    display(df)

else:
    print("❌ لم يتم رفع صورة.")
```

```python
%%writefile app.py
"""
Project: Smart Parking & ANPR System
Institution: Al-Aqaba University for Technology (AUT)
Faculty: Artificial Intelligence and Data Engineering
Supervised By: Dr. Hassan Al-Rashaideh
Created By: Ahmed Wahed Abdel Aziz & Saber Gomaa Saber Oweis
Date: April 2026
"""

import streamlit as st
from ultralytics import YOLO
import easyocr
import cv2
import sqlite3
import numpy as np
from datetime import datetime
import pandas as pd
from PIL import Image

# ==========================================
# 1. إعدادات الصفحة الأساسية
# ==========================================
st.set_page_config(page_title="نظام المواقف الذكية", page_icon="🚗", layout="wide")
st.title("🚗 النظام الذكي للتعرف على لوحات السيارات (ANPR)")
st.markdown("### جامعة العقبة للتكنولوجيا - إشراف د. حسن الرشايده")

# ==========================================
# 2. تحميل النماذج (بشكل ذكي لمنع إعادة التحميل)
# ==========================================
@st.cache_resource
def load_models():
    # تحميل YOLO و EasyOCR مرة واحدة فقط لتسريع الواجهة
    model = YOLO('/content/runs/detect/aut_anpr_model-3/weights/best.pt')
    reader = easyocr.Reader(['en'], gpu=True)
    return model, reader

model, reader = load_models()

# ==========================================
# 3. إعداد قاعدة البيانات
# ==========================================
def init_db():
    conn = sqlite3.connect('smart_parking.db')
    cursor = conn.cursor()
    cursor.execute('''CREATE TABLE IF NOT EXISTS entry_logs
                      (id INTEGER PRIMARY KEY AUTOINCREMENT, plate_number TEXT, entry_time TIMESTAMP)''')
    conn.commit()
    return conn

conn = init_db()

# ==========================================
# 4. تصميم الواجهة (UI)
# ==========================================
# تقسيم الشاشة إلى عمودين
col1, col2 = st.columns(2)

with col1:
    st.header("📥 رفع صورة السيارة")
    # أداة لرفع الصور بديلة عن Colab Upload
    uploaded_file = st.file_uploader("اختر صورة للسيارة...", type=["jpg", "jpeg", "png"])

    if uploaded_file is not None:
        # قراءة الصورة وعرضها
        image = Image.open(uploaded_file)
        st.image(image, caption='الصورة الأصلية', use_column_width=True)

        # تحويل الصورة لتناسب مكتبة OpenCV
        img_array = np.array(image)
        img_cv = cv2.cvtColor(img_array, cv2.COLOR_RGB2BGR)

with col2:
    st.header("⚙️ نتائج التحليل الذكي")
    if uploaded_file is not None:
        with st.spinner("جاري تحليل الصورة واستخراج اللوحة..."):
            # تشغيل نموذج YOLO
            results = model(img_cv)

            plate_found = False
            for result in results:
                boxes = result.boxes
                for box in boxes:
                    plate_found = True
                    # قص اللوحة
                    x1, y1, x2, y2 = box.xyxy[0].int().tolist()
                    cropped_plate = img_cv[y1:y2, x1:x2]

                    # عرض صورة اللوحة المقصوصة
                    st.image(cv2.cvtColor(cropped_plate, cv2.COLOR_BGR2RGB), caption="اللوحة المكتشفة", width=200)

                    # معالجة وقراءة النص
                    gray = cv2.cvtColor(cropped_plate, cv2.COLOR_BGR2GRAY)
                    enlarged = cv2.resize(gray, None, fx=2.5, fy=2.5, interpolation=cv2.INTER_CUBIC)
                    _, thresh = cv2.threshold(enlarged, 100, 255, cv2.THRESH_BINARY | cv2.THRESH_OTSU)

                    ocr_results = reader.readtext(thresh, allowlist='0123456789', paragraph=False)
                    plate_number = "".join([text for (bbox, text, prob) in ocr_results if prob > 0.4])

                    if plate_number:
                        st.success(f"✅ رقم اللوحة: {plate_number}")
                        # الحفظ في قاعدة البيانات
                        now = datetime.now().strftime("%Y-%m-%d %H:%M:%S")
                        conn.cursor().execute("INSERT INTO entry_logs (plate_number, entry_time) VALUES (?, ?)", (plate_number, now))
                        conn.commit()
                    else:
                        st.warning("⚠️ تم اكتشاف اللوحة لكن الأرقام غير واضحة.")

            if not plate_found:
                st.error("❌ لم يتمكن النموذج من العثور على لوحة في هذه الصورة.")

# ==========================================
# 5. عرض قاعدة البيانات في أسفل الصفحة
# ==========================================
st.markdown("---")
st.header("📊 سجل دخول السيارات (Database)")
df = pd.read_sql_query("SELECT * FROM entry_logs ORDER BY id DESC", conn)
st.dataframe(df, use_container_width=True)
```
